### Calculando e Plotando a mdia obtida para cada uma das linguagens de programao

```python
#Obtendo e plotando mdia de habilidades por linguagem
languages_names = list(languages_df.columns)

languages_mean_dict = {}
for name in languages_names:
    language_mean = languages_df[name].to_numpy().mean()
    languages_mean_dict[name] = language_mean
    
my_df = pd.DataFrame(languages_mean_dict.items())

sns.set_theme(rc={'figure.figsize':(11.7,8.27)})


ax = sns.barplot(x=0, y=1, data=my_df)
ax.set(xlabel='Linguagens', ylabel='Mdia', title='Mdia de Habilidade por Linguagem de Programao')
```

```python
#Discretizao de habilidades
number_bins = 3
languages_disct_df = languages_df.copy()

for name in languages_names:
    languages_disct_df[name] = pd.qcut(languages_disct_df[name], number_bins, ["Baixa", "Mdia", "Alta"])
    
languages_disct_df.head()
```

```python
plot_df = pd.DataFrame(languages_disct_df)

# Contagem das ocorrncias de cada categoria para cada linguagem
count_df = plot_df.apply(lambda x: x.value_counts()).T.fillna(0)
count_df.head()


# Plotagem do grfico de barras
fig, ax = plt.subplots(figsize=(12, 8))

categories = ['Alta', 'Mdia', 'Baixa']
bar_width = 0.2
index = np.arange(len(count_df))

for i, category in enumerate(categories):
    ax.bar(index + i * bar_width, count_df[category], bar_width, label=category)

# Configuraes do grfico
ax.set_xlabel('Linguagem de Programao')
ax.set_ylabel('Ocorrencias')
ax.set_title('Nvel de Habilidade por Linguagem de Programao')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(count_df.index, rotation=45)
ax.legend()

# Mostrar o grfico
plt.tight_layout()
plt.show()
```
