### Programming Language Skills Dataset

```python
column_names = list(df.columns)
cols_to_drop = column_names[11:]

languages_df = df.copy()
for col in cols_to_drop:
    languages_df = languages_df.drop(col, axis=1)
    
languages_df.head()
```

#### Defining function to add new column(s) to generate a new dataset

```python
def add_columns(column_name, used_df):
    new_df = used_df.copy()
    new_df[column_name] = df[column_name]
    
    return new_df
```

##### Adding 'Internship_Experience' column

```python
internship_df = add_columns('Internship_Experience', languages_df).copy()
internship_df.head()
```

##### Adding 'Leadership_Experience' column

```python
leadership_df = add_columns('Leadership_Experience', languages_df).copy()
leadership_df.head()
```

#### ReliefF algorithm application
Defining function to execute ReliefF algorithm 

```python
def reliefF(n_features, used_df, predict_column):
    selector = ReliefFSelector(n_features=n_features)
    # Armazenando em um array as colunas ordenadas com base na relevncia para a predio
    selected_cols = selector.select(used_df, predict_column)

    # # Criando uma nova base de dados a partir da base original
    new_df = used_df.copy().reindex(columns=selected_cols)
    return new_df
```

##### Using reliefF function to determine wich 3 languages are most relevant to predict the Internship Experience

```python
intern_reliefF_df = reliefF(3, internship_df, 'Internship_Experience').copy()
# intern_reliefF_df.drop('Internship_Experience')

intern_reliefF_df.head()
```

##### Using reliefF function to determine wich 3 languages are most relevant to predict the Leadership Experience

```python
leader_reliefF_df = reliefF(3, leadership_df, 'Leadership_Experience').copy()
leader_reliefF_df.head()
```

#### Calculating and Plotting the obtained mean for each programming language 

```python
#Obtendo e plotando mdia de habilidades por linguagem
languages_names = list(languages_df.columns)

languages_mean_dict = {}
for name in languages_names:
    language_mean = languages_df[name].to_numpy().mean()
    languages_mean_dict[name] = language_mean
    
my_df = pd.DataFrame(languages_mean_dict.items())

sns.set_theme(rc={'figure.figsize':(11.7,8.27)})


ax = sns.barplot(x=0, y=1, data=my_df)
ax.set(xlabel='Linguagens', ylabel='Mdia', title='Mdia de Habilidade por Linguagem de Programao')
```

#### Discretization of programming skills values
Categorizing numeric values by 'Low', 'Medium' and 'High'

```python
#Discretizao de habilidades
number_bins = 3
languages_disct_df = languages_df.copy()

for name in languages_names:
    languages_disct_df[name] = pd.qcut(languages_disct_df[name], number_bins, ["Low", "Medium", "High"])
    
languages_disct_df.head()
```

```python
plot_df = pd.DataFrame(languages_disct_df)

# Contagem das ocorrncias de cada categoria para cada linguagem
count_df = plot_df.apply(lambda x: x.value_counts()).T.fillna(0)
count_df.head()


# Plotagem do grfico de barras
fig, ax = plt.subplots(figsize=(12, 8))

categories = ['Low', 'Medium', 'High']
bar_width = 0.2
index = np.arange(len(count_df))

for i, category in enumerate(categories):
    ax.bar(index + i * bar_width, count_df[category], bar_width, label=category)

# Configuraes do grfico
ax.set_xlabel('Linguagem de Programao')
ax.set_ylabel('Ocorrencias')
ax.set_title('Nvel de Habilidade por Linguagem de Programao')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(count_df.index, rotation=45)
ax.legend()

# Mostrar o grfico
plt.tight_layout()
plt.show()
```

##### Getting the average value for general Hard Skills

```python
def get_mean(column_name, used_df):
    used_df[column_name] = used_df.mean(axis=1)
    print(used_df[column_name].max())
    print(used_df[column_name].min())
    
    return used_df

hardskills_df = get_mean('Hard Skills Mean', hardskills_df).copy()
hardskills_df.head()
```

##### Discretizing the means of Hard Skills

```python
def discretize(number_bins, used_df, column_name, labels):
    new_df = used_df.copy()
    new_df[column_name] = pd.qcut(new_df[column_name], number_bins, labels)
    return new_df

hardskills_disct_df = discretize(3, hardskills_df, 'Hard Skills Mean', ["Low", "Medium", "High"])   
hardskills_disct_df.head()
```
