#### ReliefF algorithm application
Defining function to execute ReliefF algorithm 

```python
def reliefF(n_features, used_df, predict_column):
    selector = ReliefFSelector(n_features=n_features)
    # Armazenando em um array as colunas ordenadas com base na relevância para a predição
    selected_cols = selector.select(used_df, predict_column)

    # # Criando uma nova base de dados a partir da base original
    new_df = used_df.copy().reindex(columns=selected_cols)
    return new_df
```

##### Using reliefF function to determine wich 3 languages are most relevant to predict the Internship Experience

```python
intern_reliefF_df = reliefF(3, internship_df, 'Internship_Experience').copy()
# intern_reliefF_df.drop('Internship_Experience')

intern_reliefF_df.head()
```

#### Discretization of programming skills values
Categorizing numeric values by 'Low', 'Medium' and 'High'

```python
#Discretização de habilidades
number_bins = 3
languages_disct_df = languages_df.copy()

for name in languages_names:
    languages_disct_df[name] = pd.qcut(languages_disct_df[name], number_bins, ["Low", "Medium", "High"])
    
languages_disct_df.head()
```

```python
plot_df = pd.DataFrame(languages_disct_df)

# Contagem das ocorrências de cada categoria para cada linguagem
count_df = plot_df.apply(lambda x: x.value_counts()).T.fillna(0)
count_df.head()


# Plotagem do gráfico de barras
fig, ax = plt.subplots(figsize=(12, 8))

categories = ['Low', 'Medium', 'High']
bar_width = 0.2
index = np.arange(len(count_df))

for i, category in enumerate(categories):
    ax.bar(index + i * bar_width, count_df[category], bar_width, label=category)

# Configurações do gráfico
ax.set_xlabel('Linguagem de Programação')
ax.set_ylabel('Ocorrencias')
ax.set_title('Nível de Habilidade por Linguagem de Programação')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(count_df.index, rotation=45)
ax.legend()

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

#### Splitting the set into 80% for training and 20% for testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### KNN Training

#### Testing KNN with 3 neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

# RODANDO KNN COM 3 VIZINHOS (NÃO DEU TEMPO DE RODAR COM MAIS) 
knn_model = KNeighborsClassifier(n_neighbors=3)# fazer um tunning talvez???#5

# TRAINANDO COM OS DADOS QUE FORAM SEPARADOS PARA TREINO ALI EM CIMA
knn_model.fit(x_train, y_train)
```

##### Evaluating performance with the test data that has been generated by:
1. Checking predictions
1. Checking KNN hits

```python
from sklearn.metrics import accuracy_score

y_pred = knn_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia do modelo KNN: {accuracy:.2f}")
```

##### Testing with 5 neighbors

```python
knn_model2 = KNeighborsClassifier(n_neighbors=5)

knn_model2.fit(x_train, y_train)
```

##### Evaluating performance with the test data that has been generated by:
1. Checking predictions
1. Checking KNN hits

```python
y_pred2 = knn_model2.predict(x_test)

accuracy2 = accuracy_score(y_test, y_pred2)
print(f"Acurácia do modelo KNN: {accuracy2:.2f}")
```

#### Tunning Test with 50 neighbors from SKLearn by:
1. Creating list of K for KNN
1. Creating list of cv scores
1. Perform 10-fold cross validation

```python
from sklearn.model_selection import cross_val_score

k_list = list(range(1,50,2))
cv_scores = []

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
```

##### Plotting Results with 50 neighbors

```python
MSE = [1 - x for x in cv_scores]

plt.figure()
plt.figure(figsize=(15,10))
plt.title('The optimal number of neighbors', fontsize=20, fontweight='bold')
plt.xlabel('Number of Neighbors K', fontsize=15)
plt.ylabel('Misclassification Error', fontsize=15)
sns.set_style("whitegrid")
plt.plot(k_list, MSE)

plt.show()
```

##### Finding best K

```python
best_k = k_list[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d." % best_k)
```

##### Testing KNN with 49 neighbors according to the results of the tunning

```python
knn_model_BEST = KNeighborsClassifier(n_neighbors=49)

knn_model_BEST.fit(x_train, y_train)

y_pred_BEST = knn_model_BEST.predict(x_test)

accuracy_BEST = accuracy_score(y_test, y_pred_BEST)
print(f"Acurácia do modelo KNN: {accuracy_BEST:.2f}")
```

#### Tunning Test with 100 neighbors

```python
k_list = list(range(1,100,2))
cv_scores = []

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
```
